```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pandas_datareader
import seaborn as sns

```

```python
# Combine the data into a single dataset
stock_pct_reduced = pd.read_csv("Outputs/stock_pct_reduced.csv")
indices_pct = pd.read_csv("Outputs/indices_pct_clean.csv")
stock_prices = pd.read_csv('Outputs/stock_data_closings.csv')

stock_prices.head(10)


# Combine the data into a single dataset
#stock_pct_reduced = pd.merge(stock_pct_reduced, indices_pct, how="left", on=["Open"])


# Display the data table for preview
#stock_pct_reduced()
```

```python
stock_count_list = len(stock_prices.columns)
stock_count_list
```

```python
stock_normalized = stock_prices / (stock_prices.iloc[0]) * 100
```

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of each stock
stock_mean=stock_normalized.apply(lambda x:x.mean())
stock_median=stock_normalized.apply(lambda x:x.median())
stock_var=stock_normalized.apply(lambda x:x.var())
stock_std=stock_normalized.apply(lambda x:x.std())
stock_sem=stock_normalized.apply(lambda x:x.sem())
```

```python
# Assemble the resulting series into a single summary DataFrame.
stock_summary_table=pd.DataFrame({
    "Mean":stock_mean,
    "Median":stock_median,
    "Variance":stock_var,
    "Std":stock_std,
    "Sem":stock_sem
})
stock_summary_table 
```

```python
#collect the list of all columns within the DataFrame
stock_summary_table.columns
#Low_high_stocks
```

```python
#Reorganize the columns using double brackets
High_low_stat=stock_summary_table[["Mean","Median","Variance","Std","Sem"]]

High_low_stat.head()
```

```python
#find the lowest and higher values stocks
high_values_mean=stock_summary_table["Mean"].sort_values(ascending=False).iloc[0:5,].index.to_list()
low_values_mean=stock_summary_table["Mean"].sort_values(ascending=True).iloc[0:5,].index.to_list()
high_values_median=stock_summary_table["Median"].sort_values(ascending=False).iloc[0:5,].index.to_list()
low_values_median=stock_summary_table["Median"].sort_values(ascending=True).iloc[0:5,].index.to_list()
high_values_var=stock_summary_table["Variance"].sort_values(ascending=False).iloc[0:5,].index.to_list()
low_values_var=stock_summary_table["Variance"].sort_values(ascending=True).iloc[0:5,].index.to_list()
high_values_std=stock_summary_table["Std"].sort_values(ascending=False).iloc[0:5,].index.to_list()
low_values_std=stock_summary_table["Std"].sort_values(ascending=True).iloc[0:5,].index.to_list()
high_values_sem=stock_summary_table["Sem"].sort_values(ascending=False).iloc[0:5,].index.to_list()
low_values_sem=stock_summary_table["Sem"].sort_values(ascending=True).iloc[0:5,].index.to_list()
high_values_mean
```

```python

high_and_low_summary_table=pd.DataFrame({
    "Highest mean":high_values_mean,
    "Lowest mean":low_values_mean,
    "Highest Median":high_values_median,
    "Lowest Median":low_values_median,
    "Highest var":high_values_var,
    "Lowest Var ":low_values_var,
    "Highest std":high_values_std,
    "Lowest std":low_values_std,
    "Highest Sem":high_values_sem,
    "Lowest Sem": low_values_sem
    
})
high_and_low_summary_table
```
